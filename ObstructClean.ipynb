```python
import pandas as pd
```

```python
# The path to our CSV file
file = "Resources/AWOIS_Obstructions.csv"

# Read our Obstructions data into pandas
#all_df = pd.read_csv(file, encoding = 'ISO-8859-1')

all_df = pd.read_csv(file, encoding='latin1')

all_df.head()
```

```python
# Get a list of all of our columns for easy reference
all_df.columns
```

```python
# Drop columns "depth" "sounding_type" "yearsunk" 
clean_df = all_df.drop(["DEPTH","SOUNDING_TYPE","YEARSUNK"], axis = 1, inplace=False)

clean_df.head()
```

```python
clean_df = clean_df.rename(columns = {'RECRD': 'Record #', 'VESSLTERMS': 'Vessel Terms',
                                      'FEATURE_TYPE': 'Feature Type', 'LATDEC': 'Lat', 'LONDEC': 'Lng',
                                      'GP_QUALITY': 'GP Quality'})
```

```python
clean_df.head()
```

```python
history_df = clean_df.filter(["Record #","HISTORY"], axis = 1)
history_df.head()
```

```python
clean_df.drop(["HISTORY"], axis = 1, inplace=True)
```

```python
clean_df.head()
```

```python
clean_df.to_csv("Resources/obstruct_cleaned.csv", index = False)
```

```python
history_df.to_csv("Resources/obstruct_history.csv", index = False)
```
