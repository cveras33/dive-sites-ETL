```python
import pandas as pd
```

<h4> Combining Wrecks and Obstructions Dive Site tables 

```python
wrecks_df = pd.read_csv("Resources/w_dive_sites_transformed.csv")
```

```python
wrecks_df = wrecks_df.rename(columns = {'Distance from Wreck':'Distance'})
```

```python
wrecks_df.head()
```

```python
obstructs_df = pd.read_csv("Resources/o_dive_sites_transformed.csv")
```

```python
obstructs_df = obstructs_df.rename(columns = {'Distance from Obstruction':'Distance'})
```

```python
obstructs_df.head()
```

```python
dfs = [wrecks_df, obstructs_df]
```

```python
dive_sites_df = pd.concat(dfs)
```

```python
dive_sites_df
```

```python
dive_sites_df.to_csv("Resources/final_dive_sites_transformed.csv", index = False)
```
